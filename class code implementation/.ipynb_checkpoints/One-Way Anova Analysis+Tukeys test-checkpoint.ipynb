## üéØ One-Way ANOVA  

This is a test to check if the **means of groups differ significantly**.  

### üìå Assumptions  
1. **üü¢ Normality** ‚Üí Data for each group should be normally distributed.  
2. **üîµ Homogeneity** ‚Üí Variance of data between groups should be roughly the same.  
   - If violated, use **Welch-ANOVA** instead.  
3. **üü£ Independence** ‚Üí The measurements of one group should not be influenced by another.  
   - If violated, use **Kruskal-Wallis Test** instead.  

### ‚öñÔ∏è Hypothesis  
1. **üü• Null Hypothesis (H‚ÇÄ)** ‚Üí The means of all groups are **equal**.  
2. **üü© Alternate Hypothesis (H‚ÇÅ)** ‚Üí The mean of **at least one** group **differs**.  


## 1. Generating data

````python
# Importations

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
# Generating dummy data

# Setting the random seed - reproducibility
np.random.seed(42)

# Generating three groups with different means - mean = 50/55/60 std = 5 samples = 30
group_1 = np.random.normal(loc = 50, scale = 5, size = 30)
group_2 = np.random.normal(loc = 55, scale = 5, size = 30)
group_3 = np.random.normal(loc = 60, scale = 5, size = 30)
````

````python
# Checking how concatenation works - it combines the generated data into a 1-dimensional array

np.concatenate([group_1, group_2, group_3]).shape
````

````python
# Creating a dataframe from the generated data

data = pd.DataFrame({
    "values": np.concatenate([group_1, group_2, group_3]) , 
    "groups": ["Group 1"]*30 + ["Group 2"]*30 + ["Group 3"]*30
})
data.sample(7)
````

## 2. Visualizing the data

# üìä Understanding Box Plots

Box plots (also called **box-and-whisker plots**) are used to visualize the distribution of data. Apart from detecting **outliers**, they also provide insights into:

### 1Ô∏è‚É£ Data Distribution & Spread  
   - Box plots summarize data using five key statistics:  
     - **Minimum** (smallest non-outlier value)  
     - **First Quartile (Q1)** (25th percentile)  
     - **Median (Q2)** (50th percentile)  
     - **Third Quartile (Q3)** (75th percentile)  
     - **Maximum** (largest non-outlier value)  

### 2Ô∏è‚É£ Central Tendency  
   - The **median** (Q2) is shown as a line inside the box, representing the central value of the dataset.  

### 3Ô∏è‚É£ Variability (Interquartile Range, IQR)  
   - The **Interquartile Range (IQR)** is calculated as:  
     \[
     IQR = Q3 - Q1
     \]
   - A larger IQR indicates more spread in the middle 50% of the data.  

### 4Ô∏è‚É£ Skewness & Symmetry  
   - **Symmetric data** ‚Üí Median is centered inside the box.  
   - **Right-skewed data** ‚Üí Median is closer to Q1, with a longer upper whisker.  
   - **Left-skewed data** ‚Üí Median is closer to Q3, with a longer lower whisker.  

### 5Ô∏è‚É£ Comparison Between Groups  
   - Multiple box plots can be plotted together to compare different categories or datasets.  
   - Useful for visualizing differences in distributions across groups.  

### 6Ô∏è‚É£ Presence of Clusters or Gaps  
   - Large **gaps** between quartiles or whiskers indicate areas where data is sparse.  
   - **Clusters** can be inferred when certain quartiles are compressed together.  

---
‚úÖ **Key Takeaway**: Box plots provide a compact yet powerful summary of a dataset's **spread, central tendency, skewness, and variability**, making them great for exploratory data analysis (EDA).

## a. Histogram + Bell curve

````python
# plotting the histogram - For all the data points

plt.figure(figsize = (5, 5))
plt.hist(x = data["values"], bins = 20, rwidth = 0.8, color = "blue", density = False)
plt.xlabel("values")
plt.ylabel("count")
````

````python
# Defining the function to plot the bell curves

from scipy.stats import norm
plt.figure(figsize = (12, 6))
# Defining the function to plot the distribution- using histograms
def plotDistributions(group, values, rwidth = 0.8, color_bins = "blue", color_line="red", density = True, histogram_only = False, ax = None):
    if ax is None:
        # Get the current axis if not provided
        ax = plt.gca()
    if histogram_only:
        ax.hist(values, bins = 20, rwidth=rwidth, color = color_bins, density = False, alpha = 0.6)
        ax.set_xlabel("values")
        ax.set_ylabel("count")     
    else:
        ax.hist(values, bins=20,  rwidth=rwidth, color=color_bins, density=density, alpha = 0.6)
        rng = np.arange(values.min(), values.max(), 0.1)
        ax.plot(
            rng, 
            norm.pdf(rng, values.mean(), values.std()),
            color=color_line
        )
        ax.set_xlabel("values")
        ax.set_ylabel("density")   
    ax.set_title(f"{group} data distribution")    
        
    
````

````python
# Creating subplots

with plt.style.context("seaborn-v0_8"):
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 6))
    # Flatten to handle 1D indexing
    axes = axes.flatten() 
    groups = [("group1", group_1), ("group2", group_2), ("group3", group_3)]
    for i, (group_name, group) in enumerate(groups):
        plotDistributions(group_name, group, ax=axes[i])
    plt.tight_layout()
    plt.show()  
````

## 2. Analytically

- Apart from graphical methods such as bell curve and quantile plot we can use analytical methods such as :
  
     1.  Shapiro-Wilk test
     2.  Kolmogorov-Smirnov test
     3. Anderson-Darling test
     4. D'Agostino and Pearson test

## Which Test to Use?

| **Test**              | **Best for**                              |
|-----------------------|------------------------------------------|
| **Shapiro-Wilk**      | Small datasets (<5000)                  |
| **Kolmogorov-Smirnov**| General purpose, works for large datasets |
| **Anderson-Darling**  | More robust for larger datasets         |
| **D'Agostino-Pearson** | Medium to large datasets                |

````python
# Creating a list for the groups

groups = [("group 1", group_1), ("group 2", group_2), ("group 3", group_3)]
````

## a. Shapiro wilk test - Works for small datasets(n < 5000)

````python
from scipy.stats import shapiro

for i, (group_name, group) in enumerate(groups):
    stat, p = shapiro(group)
    print(f'Shapiro-Wilk Test: Statistic={stat}, p-value={p}')
    if p > 0.05:
        print(f"{group_name} is normally distributed\n")
    else:
        print(f"{group_name} is not normally distributed\n")
````

## b. Kolomogorov-Smirgnov Test - Works for large datasets

````python
# Note:  This works for large dataset and out dataset is small hence results may not be accurate

from scipy.stats import kstest

for i, (group_name, group) in enumerate(groups):
    stat, p = kstest(group, "norm")
    print(f'Kolomogorov Smirgnov Test: Statistic={stat}, p-value={p}')
    if p > 0.05:
        print(f"{group_name} is normally distributed\n")
    else:
        print(f"{group_name} is not normally distributed\n")
````

## c. D'Agostino-Pearson Test

- Good for medium to large datasets.
- Checks for skewness and kurtosis.

````python
# Note:  This works for large dataset and out dataset is small hence results may not be accurate

from scipy.stats import normaltest

for i, (group_name, group) in enumerate(groups):
    stat, p = normaltest(group)
    print(f"D'Argostino-Pearson Test: Statistic={stat}, p-value={p}")
    if p > 0.05:
        print(f"{group_name} is normally distributed\n")
    else:
        print(f"{group_name} is not normally distributed\n")
````

## Homogenity - Variances of the groups are equal

- One of the assumptions of one-way anova is that variances between different groups are equal
- To check this we use the levene test

Levene Test Hypotheses
----------------------

**scipy.stats.levene**


1. Null Hypotheses(H0) - The variances of the different groups are equal
2. Alternate Hypotheses(Ha) - At least one group variance differs from the rest

Decision Rule
-------------

* If the p-value is greater than the chosen significance level (e.g., 0.05), we fail to reject H‚ÇÄ, meaning the assumption of equal variances holds.
* If the p-value is less than the significance level, we reject H‚ÇÄ, indicating that variances are significantly different across groups.

````python
# Levene test
from scipy.stats import levene

levene_stat, p_value = levene(group_1, group_2, group_3)
print(f"The levene stat\t = {levene_stat}\np-value\t\t = {p_value}")
````

````python
# Levene logic Hypothesis testing

alpha = 0.05
if levene_stat < alpha:
    print(f"The p_value < {alpha} hence we reject the null hypothesis/nThis means that at least one variance differs")
else:
    print(f"The p-value >= {alpha} hence we fail to reject the null hypothesis/nThis means that variances are equal hence homogenity assumption is satisfied")
````

## One Way Anova

Hypothesis
----------

- H0 => mean of group 1 = mean of group 2 = mean of group 3
- Ha => at least one mean is different

````python
# Anova test
from scipy.stats import f_oneway

# Initiliazing the anova test
F_statistic, p_value = f_oneway(group_1, group_2, group_3)
print(f"F-statistic\t = {F_statistic}\np-value\t\t = {p_value}")
````

````python
# Interpretation of the anova test

# Setting our significance level
alpha = 0.05

# Logic
if p_value < alpha:
    print("Rejection of the null hypothesis: At least one mean is different")
else:
    print("Failure to reject the null hypothesis: Means are the same")
````

## Post-Hoc Analysis

- When the Anova test is completed and we reject the null Hypothesis (Meaning at least one group mean differs) Post-Hoc analysis is used to determine which specific groups differ

Types of Post-Hoc Test
----------------------
1. Tukey's Honestly significant difference(HSD) - most commony used
2. Bonferroni Correction - Adjust significance level to control false positives
3. Scheffe's test - More conservative, good for unequal sample sizes
4. Dunn's test - Used for non-parametric ANOVA(Kruskall Wallis test)


- We will focus on  the Post-Hoc in this example

Tukey's honestly significant difference (Tukey's HSD)
-----------------------------------------------------

  `statsmodels.stats.multicomp.pairwise_tukeyhsd`

- Compares all positive pairs of groups while adjusting for mutiple comparisons
- It helps determine which groups differ significantly


Steps using python
------------------

## üìä Tukey HSD Analysis

```python
# Importing the required library
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Instantiations
tukey = pairwise_tukeyhsd(data["values"], data["groups"], alpha=0.05)

# Display results
print(tukey)

## üìä Tukey HSD Results

Multiple Comparison of Means - Tukey HSD, FWER=0.05 
=====================================================

# Plotting the results
tukey.plot_simultaneous()
plt.show()


## üéØ Interpreting Results  

### üìå Key Terms  

- **üü¢ group1 & group2** ‚Üí The two groups being compared  
- **üîµ meandiff** ‚Üí The difference between the two groups in comparison  
- **üü£ p-adj** ‚Üí Adjusted p-value (to control for multiple comparisons)  
- **üü† lower & upper** ‚Üí The confidence intervals for the mean difference  
- **üî¥ reject** ‚Üí If **True**, the difference is **statistically significant**  


## Visualizing the tukey plot interpretation

- This plot shows confidence interval for each comparison
- If the confidence intervals do not overlap, the difference is significant

## Decision Rule 
- If p-adj < 0.05(the threshhold we select), reject the null hypothesis meaning the groups are significantly different
- if p-adj >= 0.05, we fail to reject the null hypothesis meaning the difference is not statistically significant

## Note: Confidence interval overlap

Example of a confidence interval overlap
----------------------------------------
Group 1: (45, 55)  

Group 2: (50, 60)  

Example of a confidence interval non-overlap
--------------------------------------------
Group 1: (45, 55)

Group 2: (60, 70)

````python
# Tukey HSD Analysis
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Instantiations
tukey = pairwise_tukeyhsd(data["values"], data["groups"], alpha = 0.05)
print(tukey)
````

````python
# Plotting the results
tukey.plot_simultaneous()
plt.show()
````
