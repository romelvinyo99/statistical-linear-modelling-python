## --------------------DB-Scan Clustering---------------------------------

```python
# Importations


from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons, make_circles
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams["figure.figsize"] = (17, 6)
```

```python
# Creating the circles dataset

X, y = make_circles(
    n_samples = 750,
    shuffle = True,
    noise = 0.1,
    factor = 0.3,
    random_state = 42
)
X[1], y[1]
```

```python
# Plotting the circle dataset - non-linear data

plt.scatter(X[:, 0], X[:, 1], c = y)
```

```python
# Initializing the DB-SCAN clustering algorithm

dbscan_alg = DBSCAN(eps=0.10)
# Fitting the data - remeber we are only fitting the feature variables(X) and not the labels(y)
predictions = dbscan_alg.fit_predict(X)
# When a value is -1 --> That is an outlier
predictions[:5]
```

```python
# Plotting the data with respect to the dbscan algorithm labels

plt.scatter(X[:, 0], X[:, 1], c = dbscan_alg.labels_)
```
