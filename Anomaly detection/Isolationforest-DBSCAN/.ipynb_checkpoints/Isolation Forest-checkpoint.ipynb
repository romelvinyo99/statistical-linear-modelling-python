## -------------Isolation Forest------------------

```python
# Importations

import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np

matplotlib.rcParams["figure.figsize"] = (17, 3)
```

```python
# Loading dataset

df = pd.read_csv("weight-height.csv").drop(columns = ["Gender"], axis=1)
df.head(3)
```

```python
# Plotting weight against height

plt.scatter(x = df.iloc[:, 0], y = df.iloc[:, 1])
plt.xlabel("height")
plt.ylabel("weight")
```

```python
# Isolation forest(Unsupervised) - Fitting the dataframe
from sklearn.ensemble import IsolationForest
# Set contamination - range(0, 0.5) - amount of outliers/proportion
iso_alg = IsolationForest(contamination = 0.2)
iso_alg.fit(df)
```

```python
# Making predictions ---> -1 outliers 1 normal

predictions = iso_alg.predict(df)
predictions
```

```python
# Getting index of outliers

outlier_index = np.where(predictions < 0)
normal_index = np.where(predictions > 0)
```

```python
# Plottling the outliers

x = df.values
plt.scatter(x[outlier_index, 0], x[outlier_index, 1], edgecolors = "red", label = "outliers datapoints")
plt.scatter(x[normal_index, 0], x[normal_index, 1], edgecolors = "blue", label = "normal datapoints")
plt.xlabel("height")
plt.ylabel("weight")
plt.legend()
```
