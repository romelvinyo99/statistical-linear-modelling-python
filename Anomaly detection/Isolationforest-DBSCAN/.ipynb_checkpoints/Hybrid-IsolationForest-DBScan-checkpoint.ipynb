## ------------Concurrent Implementation of Isolation Forest and DB-Scan algorithm----------------

```python
# Importations

from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest
from sklearn.cluster import DBSCAN
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
%matplotlib inline
matplotlib.rcParams["figure.figsize"] = (17, 4)
```

```python
# Samole dataset - dataset with 1000 samples and three features

# Generating the random data
data = np.random.rand(1000, 2)
# Conversion to a pandas dataframe
data = pd.DataFrame(data)
data.head(3)
```

```python
# Scaling data --> mean = 0 and std = 1

# Initializing the scaler
scaler = StandardScaler()
# Transforming the data
scaled_data = scaler.fit_transform(data)
# Converting the scaled data to a pandas dataframe
scaled_data = pd.DataFrame(scaled_data)
scaled_data.head(3)
```

```python
# Getting summary statistics

scaled_data.describe()
```

```python
# Plotting the data

plt.scatter(x = scaled_data[0], y = scaled_data[1])
```

```python
# Anomaly detection with isolation forest algorithm

iso_alg = IsolationForest(contamination = 0.1, random_state = 42)
scaled_data["isolation"] = iso_alg.fit_predict(scaled_data)
scaled_data.sample(10)
```

```python
# Filtering out anomalies

non_anomalous_data = np.array(scaled_data[~(scaled_data["isolation"]==-1)])
anomalous_data = np.array(scaled_data[(scaled_data["isolation"]==-1)])
```

```python
# Visualizing the anomalies

plt.scatter(x = non_anomalous_data[:, 0], y = non_anomalous_data[:, 1], edgecolors = "green", color = "red")
plt.scatter(x = anomalous_data[:, 0], y = anomalous_data[:, 1], edgecolors = "blue")
```

```python
non_anomalous_data_df = pd.DataFrame(non_anomalous_data).drop(columns = [2], axis = 1)
non_anomalous_data_df.head(3)
```

```python
# Applying the DB-Scan clustering algorithm

dbscan = DBSCAN(eps = 0.1, min_samples = 3)
dbscan.fit(non_anomalous_data_df)
```

```python
# Getting the labels -- if label is -1 that is an outlier

dbscan.labels_
```

```python
# Removing the outliers

outlier_index = np.where(dbscan.labels_ == -1)
normal_index = np.where(dbscan.labels_ != -1)
```

```python
# Converting the dataframe for non anomalous data for better visualization

non_anomalous_data_df = np.array(non_anomalous_data_df)
```

```python
plt.scatter(x = non_anomalous_data_df[outlier_index, 0], y = non_anomalous_data_df[outlier_index, 1], edgecolors = "red", label = "outliers")
plt.scatter(x = non_anomalous_data_df[normal_index, 0], y = non_anomalous_data_df[normal_index, 1], edgecolors = "green", label = "normal datapoints")
plt.legend(prop = {"size": 6})
```
