```python
# Importations

import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams["figure.figsize"] = (12, 6)
```

```python
# Loading dataset

df = pd.read_csv("weight-height.csv").drop(columns = ["Gender"], axis = 1)
df.head(3)
```

```python
# Scaling data

scaled_data = StandardScaler().fit_transform(df)
scaled_data[:5]
```

```python
# Principle component analysis

components = scaled_data.shape[1]
pca = PCA()
pca_data = pca.fit_transform(scaled_data)
pca_data
```

```python
# Percentage variance ratio

per_var = np.round(pca.explained_variance_ratio_ * 100, decimals = 2)
per_var
```

```python
labels = ["PC" + str(x) for x in range(1, len(per_var)+1)]
labels
```

```python
# Drawing the scree plot

plt.bar(x = range(1, len(per_var)+1), height = per_var, tick_label = labels)
plt.title("Scree Plot")
plt.xlabel("Principal Component")
plt.ylabel("Percentage variance ratio")
plt.show()
```

```python
# Print the PCA components
pd.DataFrame(pca.components_, columns=df.columns, index=["PC1", "PC2"])
```

```python
# Create a DataFrame for the PCA data
pca_df = pd.DataFrame(pca_data, columns=["PC1", "PC2"])
```

```python
## Drawing a scatter plot
plt.scatter(pca_df.PC1, pca_df.PC2)
plt.title("PCA Scatter Plot")
plt.xlabel(f"PC1 - {per_var[0]}%")
plt.ylabel(f"PC2 - {per_var[1]}%")
plt.show()
```
