```python
from scipy.stats import norm

# Plotting the histogram
plt.hist(df.height, bins=20, rwidth=0.8)
plt.xlabel("Height(inches)")
plt.ylabel("count")
# Plotting the bell curve
rng = np.arange(df["height"].min(), df["height"].max(), 0.1)
plt.plot(
    rng,
    norm.pdf(
        rng,
        df["height"].mean(),
        df["height"].std()
            )
        )
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Histogram
plt.hist(df['height'], bins=20, rwidth=0.8, alpha=0.6, color='blue', density = True)  # Use `density=True` for normalized frequencies
plt.xlabel("Height (inches)")
plt.ylabel("Density")

# Bell curve
rng = np.arange(df['height'].min(), df['height'].max(), 0.1)
plt.plot(
    rng,
    # The pdf at each x-value - (value, mean, standard deviation )
    norm.pdf(rng, df['height'].mean(), df['height'].std()),
    color='red'
)
plt.show()

```

```python
# Creating the limmits

std_thresh = 3
upper_limit = df["height"].mean() + std_thresh * df["height"].std()
lower_limit = df["height"].mean() - std_thresh * df["height"].std()
upper_limit, lower_limit
```

```python
# Removing outliers

new_df = df[~((df["height"] < lower_limit) | (df["height"] > upper_limit))]
```

```python
removed_samples =  len(df) - len(new_df)
removed_samples
```

## Using Z-score

Z-score and the above standard deviation technique are inter-changeable

It calculates how many standard deviations away a data point is from the mean

For example if the mean is 66.37 and std = 3.84

If a value of a data point is 77.91 then the z-score for that is 3 because it is 3 standard deviations away (77.91 = 66.37 + 3 * 3.84)

Formula for getting the z-score: 

              Z = (Data-point - mean ) / standard deviation

```python
# Creating a new column for the z-scores of the data points

df["z-score"] = (df["height"] - df["height"].mean()) / df["height"].std()
df.head(3)
```

```python
# Lets say our thresh-hold is 3 standard deviations

std_thresh = 3
new_df = df[~(df["z-score"] > 3)]
new_df.head(3)
```
